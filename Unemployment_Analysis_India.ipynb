## 10. Exploratory Data Analysis (EDA)

Perform exploratory data analysis to understand the distribution of unemployment rates and identify important patterns in the data.

### Distribution of Estimated Unemployment Rate

This histogram shows how unemployment rates are distributed across all observations in the dataset. It helps identify the most common unemployment range, the overall spread of the data, and the presence of extreme values (outliers).

```python
plt.figure(figsize=(8,5))
sns.histplot(data=df1, x="Estimated Unemployment Rate (%)", bins=20, kde=True, color="skyblue")
plt.title("Distribution of Estimated Unemployment Rate", fontsize=14)
plt.xlabel("Estimated Unemployment Rate (%)")
plt.ylabel("Number of Observations")
plt.grid(alpha=0.3)
plt.show()
```

### Urban vs Rural Unemployment Rate

This box plot compares unemployment rates between urban and rural areas. It highlights differences in median unemployment, variability, and the presence of outliers in each area.

```python
plt.figure(figsize=(8,6))
sns.boxplot(data=df1, x="Area", y="Estimated Unemployment Rate (%)")
plt.title("Urban vs Rural Unemployment Rate")
plt.show()
```

### Average Unemployment Rate by Area

This bar chart compares the average unemployment rate in urban and rural regions. It provides a simple overview of which area experienced higher unemployment on average.

```python
area_avg = df1.groupby("Area")["Estimated Unemployment Rate (%)"].mean()
plt.figure(figsize=(6,5))
area_avg.plot(kind="bar", color=["skyblue","orange"])
plt.title("Average Unemployment Rate by Area")
plt.xlabel("Area")
plt.ylabel("Average Unemployment Rate (%)")
plt.xticks(rotation=0)
plt.show()
```

### Average Monthly Unemployment Rate

This line chart illustrates how the average unemployment rate changed over time. It helps identify monthly trends and major changes throughout the study period.

```python
monthly_unemployment = (df1.groupby('Date')['Estimated Unemployment Rate (%)'].mean().reset_index())
plt.figure(figsize=(14,6))
plt.plot(monthly_unemployment['Date'], monthly_unemployment['Estimated Unemployment Rate (%)'], marker='o', linewidth=2)
plt.title("Average Monthly Unemployment Rate")
plt.xlabel("Date")
plt.ylabel("Average Unemployment Rate (%)")
plt.grid(True)
plt.show()
```

### Impact of COVID-19 on Unemployment Rate

This visualization highlights the effect of the COVID-19 lockdown on unemployment by marking the lockdown period on the monthly unemployment trend. It helps demonstrate how unemployment changed before and after the lockdown.

```python
plt.figure(figsize=(14,6))
plt.plot(monthly_unemployment['Date'], monthly_unemployment['Estimated Unemployment Rate (%)'], marker='o', linewidth=2, label='Average Unemployment Rate')
plt.axvline(pd.to_datetime('2020-03-25'), color='red', linestyle='--', label='COVID-19 Lockdown')
plt.title("Impact of COVID-19 on Unemployment Rate")
plt.xlabel("Date")
plt.ylabel("Average Unemployment Rate (%)")
plt.legend()
plt.grid(True)
plt.show()
```

### Average Monthly Employment

This line chart shows the average number of employed individuals each month. It helps analyze employment trends and observe how employment levels changed over time.

```python
monthly_emp = df1.groupby(df1["Date"].dt.to_period("M"))["Estimated Employed"].mean()
monthly_emp.index = monthly_emp.index.astype(str)
plt.figure(figsize=(12,6))
plt.plot(monthly_emp.index, monthly_emp.values, marker="o")
plt.title("Average Monthly Employment")
plt.xlabel("Month")
plt.ylabel("Estimated Employed")
plt.xticks(rotation=45)
plt.show()
```

### Average Monthly Labour Participation Rate

This visualization displays the monthly average labour participation rate, showing how workforce participation varied over time during the study period.

```python
monthly_labour = df1.groupby(df1["Date"].dt.to_period("M"))["Estimated Labour Participation Rate (%)"].mean()
monthly_labour.index = monthly_labour.index.astype(str)
plt.figure(figsize=(12,6))
plt.plot(monthly_labour.index, monthly_labour.values, marker="o", color="green")
plt.title("Average Monthly Labour Participation Rate")
plt.xlabel("Month")
plt.ylabel("Participation Rate (%)")
plt.xticks(rotation=45)
plt.show()
```

### Top 10 States with Highest Average Unemployment Rate

This horizontal bar chart presents the ten states with the highest average unemployment rate. It helps identify regions that experienced relatively higher unemployment during the observation period.

```python
state_unemployment = (df1.groupby('Region')['Estimated Unemployment Rate (%)'].mean().sort_values(ascending=False))
state_unemployment.head(10)
```

```python
top10 = state_unemployment.head(10)
plt.figure(figsize=(12,6))
top10.plot(kind='barh', figsize=(10,6))
plt.title("Top 10 States with Highest Average Unemployment Rate")
plt.xlabel("State")
plt.ylabel("Average Unemployment Rate (%)")
plt.xticks(rotation=45)
plt.show()
```

### Top 10 States by Average Employment

This horizontal bar chart shows the ten states with the highest average employment levels. It highlights regions with the largest employed population.

```python
employment = df1.groupby("Region")["Estimated Employed"].mean()
top10_emp = employment.sort_values(ascending=False).head(10)
plt.figure(figsize=(12,6))
top10.plot(kind="barh")
plt.title("Top 10 States by Average Employment")
plt.xlabel("State")
plt.ylabel("Estimated Employed")
plt.xticks(rotation=45)
plt.show()
```

### Top 10 States by Labour Participation Rate

This horizontal bar chart displays the ten states with the highest average labour participation rate. It provides insight into regions where a larger proportion of the population actively participated in the workforce.

```python
labour = df1.groupby("Region")["Estimated Labour Participation Rate (%)"].mean()
top10_labour = labour.sort_values(ascending=False).head(10)
plt.figure(figsize=(12,6))
plt.barh(top10_labour.index, top10_labour.values, color="teal")
plt.title("Top 10 States by Labour Participation Rate")
plt.xlabel("Average Labour Participation Rate (%)")
plt.ylabel("State")
plt.gca().invert_yaxis()
plt.grid(axis="x", alpha=0.3)
plt.show()
```

### Correlation Heatmap

The correlation heatmap illustrates the relationships among the numerical variables in the dataset. It helps identify positive, negative, or weak correlations between unemployment rate, employment, and labour participation rate.

```python
plt.figure(figsize=(8,6))
sns.heatmap(df1.select_dtypes(include=['number']).corr(), annot=True, cmap="coolwarm", fmt=".2f")
plt.title("Correlation Heatmap")
plt.show()
```

# Conclusion

This exploratory data analysis examined unemployment trends in India using multiple visualizations and statistical summaries.

### Key Findings

- The unemployment rate distribution is right-skewed, with most observations concentrated at lower unemployment levels and a few extreme values.
- Urban areas generally experienced higher unemployment than rural areas during the study period.
- Average unemployment remained relatively stable before the COVID-19 pandemic but increased sharply after the nationwide lockdown, indicating a significant economic impact.
- Employment levels and labour participation rates also showed noticeable changes during the pandemic period.
- Certain states consistently recorded higher unemployment rates than others, highlighting regional differences in employment conditions.
- The correlation analysis showed a weak negative relationship between unemployment and employment, while labour participation rate had very little correlation with the other numerical variables.

### Overall

The analysis demonstrates that the COVID-19 pandemic had a substantial impact on unemployment across India. Regional variations, urban–rural differences, and monthly trends provide valuable insights into labour market dynamics. These findings can support policymakers and researchers in understanding employment patterns and planning strategies for economic recovery.
